```python
import seaborn as sns
import numpy as np
```

# seaborn works well with pandas so we would install pandas data set first , we have a ffunction that is present in sns.

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.barplot(x='sex' , y='total_bill' , data=tips , estimator = np.std)
# It would give us the mean of the total_bill of male and female , if estimator was not given , here in eastimator it is taking
# the standard deviation
```

# Count plot - Its agreegation function is count as we changed the aggregation function in case of bargraph from mean to standard deviation , so we need not give y as a parameter

```python
sns.countplot(x='sex' , data = tips) # It will count how many males and females were there
```

```python
sns.countplot(x='time' , data = tips)
```

# Boxplot

```python
sns.boxplot(x ='day',y='total_bill',data = tips , hue='sex')
```

# Voilin plot

```python
sns.violinplot(x='day' , y = 'total_bill' , data = tips , hue='sex')
```

```python
sns.violinplot(x='day' , y = 'total_bill' , data = tips , hue='sex' , split = True)
```

# Distribution Plot

```python
sns.distplot(tips['total_bill'])
# If we want the line (KDE - Kernel Density Estimation) i.e only histogram to disapper then
```

```python
# If we want the line (KDE - Kernel Density Estimation) i.e only histogram to disapper then
sns.distplot(tips['total_bill'], kde = False)
```

# KDE - Kernel Density Estimation Graph

```python
sns.kdeplot(tips['total_bill'])
```

# Joint Plot - We can combine two different distribution plot into one graph

```python
sns.jointplot(x='total_bill' , y='tip' , data = tips)
```

# We can also change the type of graph that we want to be plotted , as if we want hexagon then , by default its scatter 

```python
sns.jointplot(x='total_bill' , y='tip' , data = tips , kind = 'hex')
sns.jointplot(x='total_bill' , y='tip' , data = tips , kind = 'reg') # for regression
```

# Pairplot - For plotting pairwise relationship in dataset

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex')
```

```python
tips_correlation = tips.corr()
print(tips_correlation)
```

# Heat map - It should have variable on both rows and columns , light color means high correlation , dark color means low correlation

```python
sns.heatmap(tips_correlation , annot = True)
```
