```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
# REPRESENTING THE DATA ON GRAPH
plt.plot(x,y1)
plt.show()
```

```python
# CHANGING THEME IN GRAPH
themes = plt.style.available
print(themes)
plt.style.use("seaborn")
```

```python
plt.plot(x,y1,color='red',label="APPLE",linestyle = "dashed")
plt.plot(x,y2,color='green',label = "KIWI" , marker = "o")
plt.xlabel("TIME")
plt.ylabel("PRICE")
plt.legend()
plt.show()
```

# BAR GRAPH

```python
x_cor = np.array([0,1,2])*2
plt.bar(x_cor,[10,20,15],width = 0.5,label = "Current Year") # CURRENT YEAR 
plt.bar(x_cor+0.5,[20,10,5], width = 0.5,label = "Previous Year") # PREVIOUS YEAR
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_cor = np.array([0,1,2])*2
plt.bar(x_cor,[10,20,15],width = 0.5,label = "Current Year",color="blue") # CURRENT YEAR 
plt.bar(x_cor+0.5,[20,10,5], width = 0.5,label = "Previous Year",color="pink") # PREVIOUS YEAR
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
subjects = "Maths","Chem","Physics","English" 
weightage = [20,10,15,5]

plt.pie(weightage,labels=subjects)
plt.show()

plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0)) # It will show the Physics i.e. the 4rth one separate from the pie chart 
plt.show()
```

```python
plt.pie(weightage,labels=subjects,explode=(0,0,0.2,0),autopct = '%1.1f%%') # This autopct will show the percentage of each subject also 
plt.show()
```

# Normal Distribution - A random variable x{x-axis} has what probability p(x){y-axis} , mean is at middle in curve , mean is suppose 60 and standard deviation is suppose 5 , SD means majority of outcomes are in between 55 to 65

# Standard Normal Deviation - Its a normal diatribution whose mean is 0 and its standard deviation is 1 . We can make more curves from it , suppose we want to make a curve that has mean of 60 , then we will add 60 to Standard NOrmal Devaition i.e shift the curve . 

## X(Standard Normal) = {X(in extended curve)-mean(of extended curve)} / Standard deviation(of extended curve)

# X = (Xsn * SD) + mean

# We will make histogram , for normal distribution

```python
Xsn = np.random.randn(100) #Return a sample (or samples) from the "standard normal" distribution
print(Xsn) # we get 100 sample points
```

```python
sigma = 8
u = 70
X = np.round(Xsn*sigma + u) # THis will give only integers
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X)
plt.xlabel("MArks Range")
plt.ylabel("Probability/Frequency count of student")
plt.show()
```

```python
# lets make for physics
X2=np.round(Xsn*5 + 40)
plt.xlabel("MArks Range")
plt.ylabel("Probability/Frequency count of student")
plt.hist(X,alpha = 0.8,label="Maths")
plt.hist(X2,label="Physics")
```
