# Normal Ditribution is a bell shaped curve where on x axis we represent a random variable x and on Y axis we represent probability of x.Probability density is maxium around centre(mean).Higher probability density means that value x has higher probability of getting selected.

# Standard normal Distribution has mean = 0 and standard deviation = 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = 5
sigma = 2

#Below will give us the deviation whose curve is shofted 5 units and standard deviation of 2. Satndard deviation shows spread
vals = u + sigma*np.random.randn(100)
print(vals)
```

# Histogram - Its different from bar graph as bar graph relates 2 variables but thr histogram realtes onle 1 . See article of histogram on wiki.

```python
plt.hist(vals,50) # 50 bars
plt.show
```

```python
# Now cahnding sigma to 0.5 , Now spraed is lowered
sigma = 1
vals = u + sigma*np.random.randn(1000)
plt.hist(vals,10)
plt.show
```

# 5 is u , 4 is u-sigma , 3 is u - 2*sigma , similarly , 6 is u+sigma , 7 is u+2*sigma

```python
# Scatter Plot
x=vals
y=np.zeros(x.shape)

plt.scatter(x,y)
plt.show # DEnsity is close to the values lieing around the mean.
```

# Probability formula for multivariant normal distribution , (X-u)' means transpose of (X-u) , || is determinant of  matrix , n is number of dimensions

# Implementation using numpy to generate multivariant data

```python
# Bivariant Case
mean = np.array([0.0,0.0]) # Mean is centered 
cov = np.array([[1,0],[0,1]]) # This will give 0 correlation i.e. neither +ve nor -ve

dist = np.random.multivariate_normal(mean,cov,500) # I want to get 500 points
print(dist.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1]) # Most of the values are centerd along 0,0 and change in x does not affect the change in y.
```

```python
# +ve correlation
mean = np.array([0.0,0.0]) # Mean is centered 
cov = np.array([[1,0.5],[0.5,1]]) # This will give +ve correlation

dist = np.random.multivariate_normal(mean,cov,500) # I want to get 500 points
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1]) # This will give +ve correlation, like straight line(tilted along line y=x.)
```

```python
# -ve correlation
mean = np.array([0.0,0.0]) # Mean is centered 
cov = np.array([[1,0.8],[0.8,1]]) # This will give +ve correlation

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]]) # This is a 2d matrix as [[sigma xx , sigma xy],[sigma yx , sigma yy]]
dist2 = np.random.multivariate_normal(mean2,cov2,500)

dist = np.random.multivariate_normal(mean,cov,500) # I want to get 500 points
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1]) # This will give +ve correlation, like straight line(tilted along line y=x.)
plt.scatter(dist2[:,0],dist2[:,1])  # Spread along y axis is more , controlled by sigma yy
plt.show()
```
