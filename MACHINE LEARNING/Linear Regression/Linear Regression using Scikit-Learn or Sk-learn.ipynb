```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression 
%matplotlib notebook
```

```python
X,Y = make_regression(n_samples=1000,n_features=2,n_informative=2,noise=10,random_state=1)
print(X.shape,Y.shape)
```

```python
# Plotting 1st feature vs Y and 2nd feature vs Y
plt.subplot(1,2,1)
plt.scatter(X[:,0],Y)
plt.subplot(1,2,2)
plt.scatter(X[:,1],Y)
```

# In the 1st feature we can see that we can relate the 1st feature with Y but 2nd feature cannot be identified with Y. So changed interactive to 2 , earlier it was 1.
# Now printing the 3d plot using matplotlib notebook.

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection = "3d")

# Creating plot
ax.scatter(X[:,0],X[:,1],Y,color = "green");
plt.title("3D Scatter Plot")
plt.show()
```

# Now we can see the 3d graph in extended manner and we can rotate and see a hyperplane , whose equation would be Y = theta(0) +  theta(1)X(1) +  theta(2)X(2) as there are only two features.Now we have to find only [theta(0)  theta(1)  theta(2)]

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y) # This makes the model learn the parameters
```

```python
model.coef_ # gives us theta1 and theta 2 , all values from theata1 ..... thaetan are clubbed into coef_
```

```python
model.intercept_ # This is theta0 , intercept only contains theta0
```

```python
model.predict([X[0],X[1]]) # We pass on the list of X[0] and X[1] , this gives is the y(predictions) i.e y0 and y1.
```

```python
Y[0],Y[1] # These are actual Y 
```

```python
model.score(X,Y) # This gives us the coefficient of determination
```

# This is around 98%
