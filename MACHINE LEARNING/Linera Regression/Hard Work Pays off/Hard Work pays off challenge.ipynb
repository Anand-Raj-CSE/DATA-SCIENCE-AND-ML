# Linear Regression

```python
def hypothesis(x,theta):
    # theta = [theta0,theta1]
    y_ = theta[0] + theta[1]*x
    return y_

def gradient(X,Y,theta):
    m = X.shape[0]
    grad = np.zeros((2,)) # Creates a numpy array with 2 elements having initial value 0.
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*X[i]
    return grad/m

def gradient_descent(X,Y,max_steps = 100,learning_rate=0.1):
    theta = np.zeros((2,))
    errors_list = []
    theta_list = [] # This is for theta updates
    for i in range(max_steps):
        # Compute Grad
        grad = gradient(X,Y,theta)
        e=Error(X,Y,theta)
        errors_list.append(e)
        # Update theta
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        # storing the values of theta during updtaes
        theta_list.append((theta[0],theta[1]))
    return theta,errors_list,theta_list

def Error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        total_error += (y_ - Y[i])**2
    return total_error/m
```

```python
theta,error_list,theta_list = gradient_descent(X,Y)
```

```python
plt.title("This shows that the algorithm is working")
plt.plot(error_list)
```

# Calculating R2 score or the coefficiennt of determination.

```python
def r2_score(Y,y_):
    # Instead of using a loop for column matrix , np.sum is efficient.
    numerator = np.sum((Y-y_)**2)
    denominator = np.sum((Y-Y.mean())**2)
    score = (1-numerator/denominator)
    return score*100

r2_score(Y,y_)
```

```python
# Loss Actually 
T0 = np.arange(-40,40,1)
T1 = np.arange(40,120,1)

print(T0)
```

```python
T0,T1 = np.meshgrid(T0,T1)
# J is the loss functrion for every T0 and T1 , T0 and T1  are in x and z plane anj J is in y axes
J = np.zeros(T0.shape)
for i in range(J.shape[0]):
    for j in range(J.shape[1]):
        y_ = T1[i,j]*X + T0[i,j]
        J[i][j] = np.sum((Y-y_)**2)/Y.shape[0] # y minus y hat square is loss and Y.shape[0] gives us the mean squared error
        
# print(J.shape)
```

```python
# Visualize the J loss 
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(T0,T1,J,cmap='rainbow')
```

```python
# drwing contour plot
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(T0,T1,J,cmap='rainbow')
```
