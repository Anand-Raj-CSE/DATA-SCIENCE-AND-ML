# Linear Regression

```python
def hypothesis(x,theta):
    # theta = [theta0,theta1]
    y_ = theta[0] + theta[1]*x
    return y_

def gradient(X,Y,theta):
    m = X.shape[0]
    grad = np.zeros((2,)) # Creates a numpy array with 2 elements having initial value 0.
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*X[i]
    return grad/m

def gradient_descent(X,Y,max_steps = 100,learning_rate=0.1):
    theta = np.zeros((2,))
    errors_list = []
    for i in range(max_steps):
        # Compute Grad
        grad = gradient(X,Y,theta)
        e=Error(X,Y,theta)
        errors_list.append(e)
        # Update theta
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
    return theta,errors_list

def Error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        total_error += (y_ - Y[i])**2
    return total_error/m
```

```python
theta,error_list = gradient_descent(X,Y)
```

```python
plt.title("This shows that the algorithm is working")
plt.plot(error_list)
```

# Calculating R2 score or the coefficiennt of determination.

```python
def r2_score(Y,y_):
    # Instead of using a loop for column matrix , np.sum is efficient.
    numerator = np.sum((Y-y_)**2)
    denominator = np.sum((Y-Y.mean())**2)
    score = (1-numerator/denominator)
    return score*100

r2_score(Y,y_)
```
