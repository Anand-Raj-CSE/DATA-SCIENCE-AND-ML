```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfx = pd.read_csv('D:/LEARNING/DATA SCIENCE/MACHINE LEARNING/Normalization of data/linearX.csv')
dfy = pd.read_csv('D:/LEARNING/DATA SCIENCE/MACHINE LEARNING/Normalization of data/linearY.csv')

x=dfx.values
y = dfy.values

plt.scatter(x,y)
```

# Normalization

```python
x2 = (x-x.mean())/x.std()
plt.scatter(x2,y,color = 'orange',label = 'Normalized Data') # It has 0 mean and 1 standard deviation
plt.scatter(x,y,label = 'Raw Data')
plt.legend()
plt.show()
```
