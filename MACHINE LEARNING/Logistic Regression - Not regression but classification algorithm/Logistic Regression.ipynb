```python
# Data genration using numpy
mean_01 = np.array([1,0.5])
cov_01 = np.array([[1,0.1],[0.1,1.2]])

mean_02 = np.array([4,5])
cov_02 = np.array([[1.2,0.1],[0.1,1.3]])

# Generating normal distribution using numpy having 500 samples
dist_01 = np.random.multivariate_normal(mean_01,cov_01,500)
dist_02 = np.random.multivariate_normal(mean_02,cov_02,500)
print(dist_01.shape)
```

# Logistic Regression Implimentation

```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

def hypothesis(X,theta):
    """
    X - entire array(m,n+1) , 1 dummy feature
    theta - np.array(n+1,1)
    """
    return sigmoid(np.dot(X,theta))
    
def error(X,y,theta):
    """
    X - (m,n+1)
    y - (m,1)
    theta - (n+1,1)
    returns a scalar value
    """
    hi = hypothesis(X,theta)
    # * will do element by element multiplication not matrix multiplication
    e = -1*np.mean((y*np.log(hi) + ((1-y)*np.log(1-hi))))
    return e
    
```

```python
a = np.linspace(-10,10,20) # 20 data points between -10 and 10
plt.plot(a,sigmoid(a))
```

```python
def gradient(X,y,theta):
    """
    X - (m,n+1)
    y - (m,1)
    theta - (n+1,1)
    returns a gradient vector = (n+1,1)
    """
    hi = hypothesis(X,theta)
    # grad =(1/m)*(y(i) - h theta(x(i)))X(i)
    grad = -np.dot(X.T,(y-hi))
    m =X.shape[0]
    return grad/m


def gradient_descent(X,y,lr=0.1,max_itr = 500):
    n=X.shape[1]
    theta = np.zeros((n,1))
    error_list = []
    for i in range(max_itr):
        err = error(X,y,theta)
        error_list.append(err)
        grad = gradient(X,y,theta)
        
        # uodate
        theta = theta - lr*grad
        
    return(theta,error_list)
```

```python
ones = np.ones((X_train.shape[0],1))
X_New_train = np.hstack((ones,X_train))
print(X_New_train.shape)
print(X_New_train)

Y_train = Y_train.reshape((-1,1)) # reshaping as 800,1
```

```python
theta,error_list = gradient_descent(X_New_train,Y_train)
```
