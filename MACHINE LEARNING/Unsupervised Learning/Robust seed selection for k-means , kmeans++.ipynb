# Our earlier implimentation got stuck in the local minima, so scikit learn impliments Robust seed Selection techinque. This algorithm chooses initial centres so that they are statistically close to the final ones.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

# using sklearn

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X,y)
```

```python
print(kmeans.cluster_centers_)
centres = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(centres[:,0],centres[:,1],marker='*',color='orange')
plt.show()
```

```python
pred = kmeans.labels_
print(pred)
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centres[:,0],centres[:,1],marker='*',color='orange')
plt.title("Prediction")
plt.show()
```
